```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 2: Load the dataset
url = "https://bit.ly/4nejNue"
df = pd.read_csv(url)
```

```python
# Step 3: Display the first 5 rows
print("ðŸ”¹ First 5 Rows of the Dataset:")
print(df.head())
```

```python
print("\nðŸ”¹ Dataset Summary with .describe():")
print(df.describe())
```

```python
print("\nðŸ”¹ Mean of each column:")
print(df.mean(numeric_only=True))
```

```python
print("\nðŸ”¹ Median of each column:")
print(df.median(numeric_only=True))
```

```python
print("\nðŸ”¹ Standard deviation of each column:")
print(df.std(numeric_only=True))
```

```python
# Step 5: Plot histograms for all features
df.hist(figsize=(12, 8), bins=10, color='skyblue', edgecolor='black')
plt.suptitle("ðŸ“Š Histograms of All Features", fontsize=16)
plt.tight_layout()
plt.show()
```
